```python
# Identify duplicate rows in the dataset
import pandas as pd
df=pd.read_csv("customer_data.csv")
duplicate_rows = df[df.duplicated()]

# Display the duplicate rows
duplicate_rows
```

```python
# Count the number of duplicate rows in the dataset
import pandas as pd
df=pd.read_csv("customer_data.csv")
duplicate_count = df.duplicated().sum()

# Display the count of duplicate rows
duplicate_count
```

```python
# Identify duplicate rows based on specific columns (e.g., 'Name' and 'Age')
import pandas as pd
df=pd.read_csv("customer_data.csv")
duplicate_rows_specific = df[df.duplicated(subset=['Name', 'Age'])]

# Display the duplicate rows based on the specified columns
duplicate_rows_specific
```

```python
# Count duplicate rows based on specific columns (e.g., 'Name' and 'Age')
import pandas as pd
df=pd.read_csv("customer_data.csv")
duplicate_count_specific = df.duplicated(subset=['Name', 'Age']).sum()

# Display the count of duplicates based on the specified columns
print("Number of duplicate rows based on 'Name' and 'Age':", duplicate_count_specific)
```

```python
# Drop duplicate rows and keep only the first occurrence
import pandas as pd
df=pd.read_csv("customer_data.csv")
df_cleaned = df.drop_duplicates()

# Display the cleaned dataset
df_cleaned
```

```python
# Drop duplicate rows based on specific columns (e.g., 'Name' and 'Age') and keep only the first occurrence
import pandas as pd
df=pd.read_csv("customer_data.csv")
df_cleaned_specific = df.drop_duplicates(subset=['Name', 'Age'])

# Display the cleaned dataset
df_cleaned_specific
```

# Task: Handling outliers

## 1. Visualize Outliers Using Box and Scatter Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Box plot for Income and PurchaseAmount

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Income'])
plt.title('Box plot of Income')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['PurchaseAmount'])
plt.title('Box plot of Purchase Amount')
plt.show()
```

### Scatter plot for Income vs PurchaseAmount

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['Income'], df['PurchaseAmount'])
plt.xlabel('Income')
plt.ylabel('Purchase Amount')
plt.title('Scatter plot of Income vs Purchase Amount')
plt.show()
```

## 2. Handle Outliers Using IQR Method

### Identify Outliers:

```python
# Calculate IQR for Income
Q1_income = df['Income'].quantile(0.25)
Q3_income = df['Income'].quantile(0.75)
IQR_income = Q3_income - Q1_income

# Calculate IQR for PurchaseAmount
Q1_purchase = df['PurchaseAmount'].quantile(0.25)
Q3_purchase = df['PurchaseAmount'].quantile(0.75)
IQR_purchase = Q3_purchase - Q1_purchase

# Define outlier boundaries
lower_bound_income = Q1_income - 1.5 * IQR_income
upper_bound_income = Q3_income + 1.5 * IQR_income

lower_bound_purchase = Q1_purchase - 1.5 * IQR_purchase
upper_bound_purchase = Q3_purchase + 1.5 * IQR_purchase

# Filter out outliers
df_iqr_removed = df[~((df['Income'] < lower_bound_income) | (df['Income'] > upper_bound_income) |
                      (df['PurchaseAmount'] < lower_bound_purchase) | (df['PurchaseAmount'] > upper_bound_purchase))]

plt.figure(figsize=(6, 6))
plt.scatter(df_iqr_removed['Income'], df_iqr_removed['PurchaseAmount'])
plt.xlabel('Income')
plt.ylabel('Purchase Amount')
plt.title('Scatter plot of Income vs Purchase Amount')
plt.show()
```

### Cap Outliers:

```python
# Cap values beyond the boundaries using .loc
df.loc[df['Income'] < lower_bound_income, 'Income'] = lower_bound_income
df.loc[df['Income'] > upper_bound_income, 'Income'] = upper_bound_income

df.loc[df['PurchaseAmount'] < lower_bound_purchase, 'PurchaseAmount'] = lower_bound_purchase
df.loc[df['PurchaseAmount'] > upper_bound_purchase, 'PurchaseAmount'] = upper_bound_purchase
```

## 3. Handle Outliers Using Z-Score Method

### Identify Outliers:

```python
from scipy import stats

# Calculate Z-scores for Income and PurchaseAmount using .assign()
df = df.assign(
    Income_zscore=stats.zscore(df['Income']),
    PurchaseAmount_zscore=stats.zscore(df['PurchaseAmount'])
)

# Alternatively, using .loc
df.loc[:, 'Income_zscore'] = stats.zscore(df['Income'])
df.loc[:, 'PurchaseAmount_zscore'] = stats.zscore(df['PurchaseAmount'])
```

### Cap Outliers:

```python
# Cap Z-scores greater than 3 or less than -3 using .loc
df.loc[df['Income_zscore'].abs() > 3, 'Income'] = df['Income'].median()
df.loc[df['PurchaseAmount_zscore'].abs() > 3, 'PurchaseAmount'] = df['PurchaseAmount'].median()
```

## 4. Handle Outliers Using Percentile Method

### Remove Outliers:

```python
# Define percentile boundaries
lower_percentile_income = df['Income'].quantile(0.01)
upper_percentile_income = df['Income'].quantile(0.99)

lower_percentile_purchase = df['PurchaseAmount'].quantile(0.01)
upper_percentile_purchase = df['PurchaseAmount'].quantile(0.99)

# Filter out outliers
df_percentile_removed = df[(df['Income'] >= lower_percentile_income) & (df['Income'] <= upper_percentile_income) &
                           (df['PurchaseAmount'] >= lower_percentile_purchase) & (df['PurchaseAmount'] <= upper_percentile_purchase)]

plt.figure(figsize=(6, 6))
plt.scatter(df_percentile_removed['Income'], df_percentile_removed['PurchaseAmount'])
plt.xlabel('Income')
plt.ylabel('Purchase Amount')
plt.title('Scatter plot of Income vs Purchase Amount')
plt.show()
```

### Cap Outliers:

```python
# Cap values to 1st and 99th percentiles using .loc with casting
df.loc[df['Income'] < lower_percentile_income, 'Income'] = float(lower_percentile_income)
df.loc[df['Income'] > upper_percentile_income, 'Income'] = float(upper_percentile_income)

df.loc[df['PurchaseAmount'] < lower_percentile_purchase, 'PurchaseAmount'] = float(lower_percentile_purchase)
df.loc[df['PurchaseAmount'] > upper_percentile_purchase, 'PurchaseAmount'] = float(upper_percentile_purchase)
```

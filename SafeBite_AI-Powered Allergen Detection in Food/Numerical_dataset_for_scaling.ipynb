```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/radhikaasmar/demo2/SafeBite_AI-Powered Allergen Detection in Food/numerical_dataset_for_scaling.csv")
df
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Dropping the 'CustomerID' column as it's not typically used in scaling
df_features = df.drop(columns=['CustomerID'])
```

```python
# 1. Normalization (MinMaxScaler)
min_max_scaler = MinMaxScaler()
df_normalized = pd.DataFrame(min_max_scaler.fit_transform(df_features), columns=df_features.columns)

print("Normalized Data:\n")
df_normalized.head()
```

```python
# 2. Standardization (StandardScaler)
standard_scaler = StandardScaler()
df_standardized = pd.DataFrame(standard_scaler.fit_transform(df_features), columns=df_features.columns)

print("\nStandardized Data:\n")
df_standardized.head()
```

```python
import joblib
import pandas as pd
import numpy as np
import os
```

```python
example_df=pd.read_csv('C:/Users/Admin/Desktop/sample.csv',na_values=[""],keep_default_na=False)
```

```python
example_df.head(9)
```

```python
# Rename the columns
example_df.rename(columns={'Price ($)': 'Price', 'Customer rating (Out of 5)': 'Customer rating'}, inplace=True)
```

```python
# Load the model and encoder from the file
encoder_path='C:/Users/Admin/Desktop/food_allergen_detection/model/encoder.pkl'
model_path = 'C:/Users/Admin/Desktop/food_allergen_detection/model/rf_model.pkl'

encoder=joblib.load(encoder_path)
rf_model = joblib.load(model_path)
```

```python
categorical_columns = ['Food Product', 'Main Ingredient', 'Sweetener', 'Fat/Oil', 'Seasoning','Allergens']

example_df_encoded=encoder.transform(example_df[categorical_columns])

example_df=pd.concat([example_df.drop(categorical_columns,axis=1),example_df_encoded],axis=1)
```

```python
# Make predictions using the loaded model
prediction = rf_model.predict(example_df)

print(prediction)
```

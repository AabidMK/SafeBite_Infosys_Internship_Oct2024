```python
import joblib
import pandas as pd
import numpy as np
import os
```

```python
example_df=pd.read_csv('sample.csv',na_values=[""],keep_default_na=False)
```

```python
example_df.head(9)
```

```python
# Load the model and encoder from the file
encoder_path='encoder.pkl'
model_path = 'rf_model.pkl'

encoder=joblib.load(encoder_path)
rf_model = joblib.load(model_path)
```

```python
categorical_columns = ['Food Product', 'Main Ingredient', 'Sweetener', 'Fat/Oil', 'Seasoning','Allergens']

example_df_encoded=encoder.transform(example_df[categorical_columns])

example_df=pd.concat([example_df.drop(categorical_columns,axis=1),example_df_encoded],axis=1)
```

```python
example_df
```

```python
import pickle

# Load the saved model
with open('rf_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

# Use the loaded model to make predictions
prediction = loaded_model.predict(example_df)

print(prediction)
```

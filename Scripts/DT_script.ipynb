# **LOGISTIC REGRESSION**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
# 70:30 Split Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_70)
X_test_scaled = scaler.transform(X_test_70)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train_70)

y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

train_accuracy = accuracy_score(y_train_70, y_train_pred)
test_accuracy = accuracy_score(y_test_70, y_test_pred)
precision = precision_score(y_test_70, y_test_pred, average='binary')
recall = recall_score(y_test_70, y_test_pred, average='binary')
f1 = f1_score(y_test_70, y_test_pred, average='binary')
conf_matrix = confusion_matrix(y_test_70, y_test_pred)
class_report = classification_report(y_test_70, y_test_pred)

# Print evaluation metrics
print("Model Evaluation Metrics:")
print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
```

# **HYPERPARAMETER TUNING RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True]
}


# Initialize the RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)

# Perform Grid Search with Cross Validation
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')
grid_search.fit(X_train_70, y_train_70)

# Get the best model from the grid search
best_rf_model = grid_search.best_estimator_

# Print the best hyperparameters found by GridSearchCV
print("Best Hyperparameters:", grid_search.best_params_)

# Training and Testing Accuracy with the Tuned Model
y_train_pred = best_rf_model.predict(X_train_70)
y_test_pred = best_rf_model.predict(X_test_70)

train_accuracy = accuracy_score(y_train_70, y_train_pred)
test_accuracy = accuracy_score(y_test_70, y_test_pred)

print("\nTuned Random Forest Model Accuracy:")
print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

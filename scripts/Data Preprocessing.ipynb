```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('Allergen_Status_of_Food_Products.csv')

# Set the plot style for better aesthetics
sns.set(style="whitegrid")

# Create a figure with subplots for the box plot and scatter plot
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Outlier Visualization for Price and Customer Rating', fontsize=16)

# Box plot for Price
sns.boxplot(ax=axes[0, 0], y=data['Price ($)'], color='skyblue')
axes[0, 0].set_title('Box Plot of Price ($)')

# Scatter plot for Price
sns.scatterplot(ax=axes[0, 1], x=range(len(data)), y=data['Price ($)'], color='purple')
axes[0, 1].set_title('Scatter Plot of Price ($)')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Price ($)')

# Box plot for Customer rating
sns.boxplot(ax=axes[1, 0], y=data['Customer rating (Out of 5)'], color='lightgreen')
axes[1, 0].set_title('Box Plot of Customer Rating (Out of 5)')

# Scatter plot for Customer rating
sns.scatterplot(ax=axes[1, 1], x=range(len(data)), y=data['Customer rating (Out of 5)'], color='orange')
axes[1, 1].set_title('Scatter Plot of Customer Rating (Out of 5)')
axes[1, 1].set_xlabel('Index')
axes[1, 1].set_ylabel('Customer Rating (Out of 5)')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import pandas as pd
from scipy.stats import zscore

data = pd.read_csv('Allergen_Status_of_Food_Products.csv')

def detect_outliers_zscore(df, column, threshold=3):
    """
    Detects outliers in a given column of the dataframe using the Z-score method.
    Returns the indices of the outliers.
    """
    z_scores = zscore(df[column])
    outliers = df[(z_scores > threshold) | (z_scores < -threshold)].index
    return outliers

# Detecting outliers in 'Price ($)' and 'Customer rating (Out of 5)'
price_outliers = detect_outliers_zscore(data, 'Price ($)')
rating_outliers = detect_outliers_zscore(data, 'Customer rating (Out of 5)')

# Handling Outliers
# Option 1: Removing outliers
data_removed_outliers = data.drop(index=price_outliers).drop(index=rating_outliers)

# Option 2: Capping outliers
data_capped = data.copy()
price_mean = data['Price ($)'].mean()
price_std = data['Price ($)'].std()
rating_mean = data['Customer rating (Out of 5)'].mean()
rating_std = data['Customer rating (Out of 5)'].std()

# Define the capping limits using the Z-score threshold
price_lower_cap = price_mean - 3 * price_std
price_upper_cap = price_mean + 3 * price_std
rating_lower_cap = rating_mean - 3 * rating_std
rating_upper_cap = rating_mean + 3 * rating_std

data_capped['Price ($)'] = data_capped['Price ($)'].clip(lower=price_lower_cap, upper=price_upper_cap)
data_capped['Customer rating (Out of 5)'] = data_capped['Customer rating (Out of 5)'].clip(lower=rating_lower_cap, upper=rating_upper_cap)

# Display the results
print("Original Data Shape:", data.shape)
print("Data Shape After Removing Outliers:", data_removed_outliers.shape)
print("Data Shape After Capping Outliers:", data_capped.shape)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('Allergen_Status_of_Food_Products.csv')

# Create a copy to preserve original data
df_imputed_mode = df.copy()

# Impute missing values using the mode for both numerical and categorical columns
for col in df_imputed_mode.columns:
    # Fill missing values with the mode (most frequent value)
    mode_value = df_imputed_mode[col].mode()[0]
    df_imputed_mode[col].fillna(mode_value, inplace=True)

# Set pandas display option to show all rows (optional)
pd.set_option('display.max_rows', None)

# Check the result
print(df_imputed_mode)
```

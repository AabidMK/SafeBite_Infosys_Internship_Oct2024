```python
# Impute missing values in numerical columns using the mean
import pandas as pd
df=pd.read_csv('Generated_Customer_Dataset.csv')
df_imputed = df.copy()  # Create a copy to preserve the original data
numerical_cols = df_imputed.select_dtypes(include=['float64', 'int64']).columns

# Impute the mean for each numerical column
for col in numerical_cols:
    df_imputed[col].fillna(df_imputed[col].mean(), inplace=True)

# Check the result
df_imputed.head()
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('Generated_Customer_Dataset.csv')

# Create a copy to preserve original data
df_imputed_mode = df.copy()

# Impute missing values using the mode for both numerical and categorical columns
for col in df_imputed_mode.columns:
    # Fill missing values with the mode (most frequent value)
    mode_value = df_imputed_mode[col].mode()[0]
    df_imputed_mode[col].fillna(mode_value, inplace=True)

# Set pandas display option to show all rows (optional)
pd.set_option('display.max_rows', None)

# Check the result
print(df_imputed_mode)
```

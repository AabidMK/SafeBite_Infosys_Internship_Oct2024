```python
import pandas as pd
import io
from sklearn.model_selection import train_test_split
df = pd.read_csv(io.BytesIO(uploaded['preprocessed_d.csv']))
```

```python
df.head()
```

```python
X=df.drop(columns=['Prediction'])
y=df['Prediction']
```

```python
X_train_70, X_test_70, y_train_70, y_test_70= train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train_70.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Parameter grid for Decision Tree
dt_param_grid = {
    'max_depth': [10],
    'min_samples_split': [10],
    'min_samples_leaf': [2],
    'criterion': ['entropy']
}

# Initialize the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)

# Perform Grid Search with Cross Validation for Decision Tree
dt_grid_search = GridSearchCV(estimator=dt_model, param_grid=dt_param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')
dt_grid_search.fit(X_train_70, y_train_70)

# Get the best model from the grid search
best_dt_model = dt_grid_search.best_estimator_

# Predictions and Accuracy (After Tuning)
y_train_pred_dt = best_dt_model.predict(X_train_70)
y_test_pred_dt = best_dt_model.predict(X_test_70)

train_accuracy_dt = accuracy_score(y_train_70, y_train_pred_dt)
test_accuracy_dt = accuracy_score(y_test_70, y_test_pred_dt)

# Print the best hyperparameters and accuracy
print("Best Decision Tree Hyperparameters:", dt_grid_search.best_params_)
print("\nTuned Decision Tree Model Accuracy:")
print("Training Accuracy:", train_accuracy_dt)
print("Testing Accuracy:", test_accuracy_dt)
```

```python
from joblib import dump, load

# Saving an object
dump(best_dt_model, "dt_model.joblib")

# Loading an object
model = load("dt_model.joblib")
```

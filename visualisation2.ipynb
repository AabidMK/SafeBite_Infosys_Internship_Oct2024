```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Allergen_Status_of_Food_Products.csv')
rating_distribution = data['Customer rating (Out of 5)'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.bar(rating_distribution.index, rating_distribution.values, color='skyblue',edgecolor='black')
plt.title('Customer Rating Distribution')
plt.xlabel('Customer rating (Out of 5)')
plt.ylabel('Number of Products')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Allergen_Status_of_Food_Products.csv')
allergen_count = data['Prediction'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(allergen_count.index, allergen_count.values, color=['lightblue', 'lightgreen'])
plt.title('Count of Products by Allergen Prediction')
plt.xlabel('Allergen Prediction')
plt.ylabel('Number of Products')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('Allergen_Status_of_Food_Products.csv')
data['Allergen Presence'] = data['Prediction'].apply(lambda x: 'Contains' if x == 'Contains' else 'Does Not Contain')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Allergen Presence', y='Price ($)', data=data, palette='Set1')
plt.title('Price Distribution by Allergen Presence')
plt.xlabel('Allergen Presence')
plt.ylabel('Price ($)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Allergen_Status_of_Food_Products.csv')
ingredient_counts = data['Main Ingredient'].value_counts()
top_10_ingredients = ingredient_counts.head(10)
plt.figure(figsize=(12, 6))
top_10_ingredients.plot(kind='bar', color='cornflowerblue')
plt.title('Top 10 Most Common Main Ingredients')
plt.xlabel('Main Ingredient')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Allergen_Status_of_Food_Products.csv')
allergens_exploded = data['Allergens'].str.split(', ').explode()
allergen_counts = allergens_exploded.value_counts()
top_10_allergens = allergen_counts.head(10)
plt.figure(figsize=(12, 6))
top_10_allergens.plot(kind='bar', color='blue')
plt.title('Top 10 Most Frequent Allergens')
plt.xlabel('Allergen')
plt.ylabel('Count')
plt.show()
```
